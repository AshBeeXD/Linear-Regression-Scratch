# Linear Regression from Scratch
This notebook demonstrates how to implement a Linear Regression model from scratch using only NumPy.

## Step 1: Implementing Linear Regression from Scratch
Define a **LinearRegressionScratch** class that:
- Initializes weights and bias
- Fits the model using gradient descent
- Makes predictions on new data
- Calculates the Mean Squared Error (MSE)

```python
import numpy as np
import matplotlib.pyplot as plt

class LinearRegressionScratch:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.weight = None
        self.bias = None

    def fit(self, X, y):
        # Initializing weights and bias
        n_samples, n_features = X.shape
        self.weight = np.zeros(n_features)
        self.bias = 0

        # Gradient descent
        for _ in range(self.n_iters):
            # Prediction
            y_pred = np.dot(X, self.weight) + self.bias

            # Computing gradients
            dw = (1/n_samples) * np.dot(X.T, (y_pred - y))
            db = (1/n_samples) * np.sum(y_pred - y)

            # Updating parameters
            self.weight -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        return np.dot(X, self.weight) + self.bias

    def mse(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)
```

## Step 2: Training the Model on Synthetic Data
Generate synthetic linear data:
\[
y = 2x + 3 + $\epsilon$
\]
where \($\epsilon$) is Gaussian noise.  

Then train the Linear Regression model on this dataset.

```python
if __name__ == "__main__":
    # Generating a synthetic dataset
    X = np.linspace(0, 10, 100).reshape(-1, 1)
    y = 2 * X.flatten() + 3 + np.random.randn(100)

    # Model training
    model = LinearRegressionScratch(learning_rate=0.01, n_iters=1000)
    model.fit(X, y)
    predictions = model.predict(X)

    print("Learned weights:", model.weight)
    print("Learned bias:", model.bias)
    print("MSE:", model.mse(y, predictions))
```

## Step 3: Visualizing the Results
Finally, plot:
- The raw data points (blue scatter)
- The best fit line predicted by our model (red line)

```python
# Plotting results
plt.scatter(X, y, color="blue", label="Data")
plt.plot(X, predictions, color="red", label="Best Fit Line")
plt.legend()
plt.show()
```
